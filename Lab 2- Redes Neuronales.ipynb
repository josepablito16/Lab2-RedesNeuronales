A continuación, se proponen dos formas de seleccionar la cantidad de neuronas en la capa oculta, que serán los dos modelos que se van a discutir: 


1.	Seleccionar un numero aleatorio entre el rango anteriormente mencionado. 
    
    Ej: 130 neuronas
